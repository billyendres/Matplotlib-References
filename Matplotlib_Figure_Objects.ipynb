```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data
a = np.linspace(0, 10, 11)
```

```python
b = a ** 4
```

```python
x = np. arange(0,10)
```

```python
y = x * 2
```

```python
fig = plt.figure()

#Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b);

axes1.set_xlim(0, 10)
axes1.set_ylim(0, 11000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('A - Power of 4')

#Small Axes
axes2 = fig.add_axes([0.15, 0.4, 0.25, 0.25])
axes2.plot(a, b);

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 20)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed Insert')
```
