```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
#Tuple Unpacking
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)

axes[1].plot(a, b)
axes[1].set_title('Axes 1')
axes[1].set_ylabel('Number')
axes[1].set_xlabel('Date')

fig.suptitle('Figure Level', fontsize=14)

#Auto Adjust
#plt.tight_layout()

#Manual Adjust
fig.subplots_adjust(wspace=0.4, hspace=None)

fig.set_figwidth(12)

fig.savefig('new_subplots.jpg', bbox_inches='tight')
```
