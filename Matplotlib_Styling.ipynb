```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X');
ax.plot(x, x**2, label='X ^ 2');
ax.legend(loc='upper left');
#ax.legend((loc=1.1, 05));
```

```python
y = np.linspace(0, 1100, 1000)
```

```python
fig_two = plt.figure()

ax_two = fig_two.add_axes([0, 0, 1, 1])

ax_two.plot(y, y, color='#FB749C');
ax_two.plot(y, y + 100, color='#74FBFA', lw= 3, ls='-.');
```
